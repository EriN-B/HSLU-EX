```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Define the n values from the table
n_values = np.array([1, 2, 5, 10, 20, 50, 100])

# Calculate each function based on n
ln_n = np.log(n_values)                # Natural logarithm ln(n)
ld_n = np.log2(n_values)               # Logarithm base 2
n_ln_n = n_values * np.log(n_values)     # n * ln(n)
n_square = n_values ** 2                 # n^2
n_cube = n_values ** 3                   # n^3
two_power_n = 2 ** n_values              # 2^n
three_power_n = 3 ** n_values            # 3^n
n_factorial = np.array([math.factorial(int(n)) for n in n_values])  # n!

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(n_values, ln_n, 'o-', label=r'$\ln(n)$')
plt.plot(n_values, ld_n, 'o-', label=r'$\log_2(n)$')
plt.plot(n_values, n_ln_n, 'o-', label=r'$n \ln(n)$')
plt.plot(n_values, n_square, 'o-', label=r'$n^2$')
plt.plot(n_values, n_cube, 'o-', label=r'$n^3$')
plt.plot(n_values, two_power_n, 'o-', label=r'$2^n$')
plt.plot(n_values, three_power_n, 'o-', label=r'$3^n$')
plt.plot(n_values, n_factorial, 'o-', label=r'$n!$')

plt.yscale("log")  # Use a logarithmic scale for the y-axis
plt.xlabel("n")
plt.ylabel("Function value (log scale)")
plt.title("Growth Functions for Selected Values of n")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.tight_layout()
plt.show()
```
